```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Simulation parameters
area_size = 10.0
road_width = 1.5
start_regions = [((0.25, 4.25), 0.5, 1.5), ((4.25, 0.25), 1.5, 0.5)]
goal_regions = [((4.25, 9.25), 1.5, 0.5), ((9.25, 4.25), 0.5, 1.5)]
dt = 0.5
spawn_rate = 0.05
num_steps = 1100  # reduced for speed

# Physical constants
v0, tau = 1.3, 0.5
A, B, r = 1.0, 1.0, 0.5
k_center = 5.0
goal_thresh = 0.5

# Constraint function
def constrain_to_roads(pos):
    if 4.25 <= pos[1] <= 4.25 + road_width:
        return np.clip(pos, 0, area_size)
    if 4.25 <= pos[0] <= 4.25 + road_width:
        return np.clip(pos, 0, area_size)
    h_proj = np.array([np.clip(pos[0], 0, area_size),
                       np.clip(pos[1], 4.25, 4.25 + road_width)])
    v_proj = np.array([np.clip(pos[0], 4.25, 4.25 + road_width),
                       np.clip(pos[1], 0, area_size)])
    return h_proj if np.linalg.norm(pos - h_proj) < np.linalg.norm(pos - v_proj) else v_proj

# Agent class
class Agent:
    def __init__(self, start_region):
        (x0, y0), w, h = start_region
        self.pos = np.array([random.uniform(x0, x0+w), random.uniform(y0, y0+h)])
        self.vel = np.zeros(2)
        gr = random.choice(goal_regions)
        (gx, gy), gw, gh = gr
        self.goal = np.array([random.uniform(gx, gx+gw), random.uniform(gy, gy+gh)])
        self.traj = [self.pos.copy()]
        self.active = True

    def force_goal(self):
        e = self.goal - self.pos
        d = np.linalg.norm(e)
        if d > 0: e /= d
        return (v0 * e - self.vel) / tau

    def force_interaction(self, others):
        if np.linalg.norm(self.pos - self.goal) < goal_thresh:
            return np.zeros(2)
        F = np.zeros(2)
        for o in others:
            if o is self:
                continue
            dv = self.pos - o.pos
            dist = np.linalg.norm(dv)
            if dist > 0:
                F += A * np.exp((2*r - dist) / B) * (dv/dist)
        return F

    def force_centering(self):
        center = 4.25 + road_width/2
        F = np.zeros(2)
        if 4.25 <= self.pos[1] <= 4.25 + road_width:
            F[1] = k_center * (center - self.pos[1])
        if 4.25 <= self.pos[0] <= 4.25 + road_width:
            F[0] = k_center * (center - self.pos[0])
        return F

    def step(self, others):
        if not self.active:
            return
        F = self.force_goal() + self.force_interaction(others) + self.force_centering()
        self.vel += F * dt
        newpos = self.pos + self.vel * dt
        self.pos = constrain_to_roads(newpos)
        self.traj.append(self.pos.copy())
        if np.linalg.norm(self.pos - self.goal) < 0.2:
            self.active = False

# Run simulation
agents = [Agent(sr) for sr in start_regions]
for _ in range(num_steps):
    for sr in start_regions:
        if random.random() < spawn_rate:
            agents.append(Agent(sr))
    new_agents = []
    for a in agents:
        a.step(agents)
        if a.active:
            new_agents.append(a)
    agents = new_agents

# Extract trajectories
trajectories = [np.array(a.traj) for a in agents if len(a.traj) > 1]

# Plot trajectories
plt.figure(figsize=(6,6))
for traj in trajectories:
    plt.plot(traj[:,0], traj[:,1], alpha=0.7)
plt.title('Collected Agent Trajectories (100 steps)')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# --- 2) 전처리: 노이즈 제거, 보간, 벡터화 ---
from sklearn.svm import OneClassSVM
from scipy.signal import savgol_filter

processed = []
for traj in trajectories:
    x_s = savgol_filter(traj[:,0], 5, 2)
    y_s = savgol_filter(traj[:,1], 5, 2)
    sm = np.vstack([x_s,y_s]).T
    idxs = np.linspace(0,len(sm)-1,16).astype(int)
    sub = sm[idxs]
    processed.append(sub.flatten())
X_train = np.vstack(processed)

print("훈련 데이터셋 크기:", X_train.shape)

# --- 3) One-Class SVM 학습 ---
model = OneClassSVM(kernel='rbf', nu=0.05, gamma='scale')
model.fit(X_train)

print("지원 벡터 개수:", len(model.support_vectors_))
print("모델 offset (ρ):", model.offset_)
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from sklearn.svm import OneClassSVM

# --- 1) 기존 SVM 학습 코드 유지 ---
area_size = 10.0
road_width = 1.5
start_regions = [((0.25,4.25),0.5,1.5), ((4.25,0.25),1.5,0.5)]
goal_regions  = [((4.25,9.25),1.5,0.5), ((9.25,4.25),0.5,1.5)]
dt, spawn_rate, num_steps = 0.5, 0.05, 2000
v0, tau, A, B, r, k_center, goal_thresh = 1.3, 0.5, 1.0, 1.0, 0.5, 5.0, 0.5

def constrain_to_roads(pos):
    if 4.25 <= pos[1] <= 4.25+road_width or 4.25 <= pos[0] <= 4.25+road_width:
        return np.clip(pos, 0, area_size)
    h = np.array([np.clip(pos[0],0,area_size), np.clip(pos[1],4.25,4.25+road_width)])
    v = np.array([np.clip(pos[0],4.25,4.25+road_width), np.clip(pos[1],0,area_size)])
    return h if np.linalg.norm(pos-h)<np.linalg.norm(pos-v) else v

class Agent:
    def __init__(self, start_region):
        (x0,y0),w,h = start_region
        self.pos = np.array([random.uniform(x0,x0+w), random.uniform(y0,y0+h)])
        self.vel = np.zeros(2)
        gr = random.choice(goal_regions)
        (gx,gy),gw,gh = gr
        self.goal = np.array([random.uniform(gx,gx+gw), random.uniform(gy,gy+gh)])
        self.traj = [self.pos.copy()]
        self.active = True
    def step(self, others):
        if not self.active: return
        # goal force
        e = self.goal - self.pos
        d = np.linalg.norm(e)
        if d>0: e/=d
        F = (v0*e - self.vel)/tau + self.interaction(others) + self.centering()
        self.vel += F*dt
        self.pos = constrain_to_roads(self.pos + self.vel*dt)
        self.traj.append(self.pos.copy())
        if np.linalg.norm(self.pos-self.goal)<0.2: self.active=False
    def interaction(self, others):
        F = np.zeros(2)
        for o in others:
            if o is self: continue
            dv = self.pos - o.pos
            dist = np.linalg.norm(dv)
            if dist>0: F += A*np.exp((2*r-dist)/B)*(dv/dist)
        return F
    def centering(self):
        center = 4.25+road_width/2
        F = np.zeros(2)
        if 4.25 <= self.pos[1]<=4.25+road_width: F[1]=k_center*(center-self.pos[1])
        if 4.25 <= self.pos[0]<=4.25+road_width: F[0]=k_center*(center-self.pos[0])
        return F

# train on pure normal
agents = [Agent(sr) for sr in start_regions]
for _ in range(int(num_steps/4)):
    for sr in start_regions:
        if random.random()<spawn_rate: agents.append(Agent(sr))
    new = []
    for a in agents:
        a.step(agents)
        if a.active: new.append(a)
    agents = new
# preprocess & flatten
def preprocess(trajs, pts=16):
    X=[]
    for t in trajs:
        if len(t)<5: continue
        xs = savgol_filter(t[:,0],5,2); ys = savgol_filter(t[:,1],5,2)
        seq = np.vstack([xs,ys]).T
        idx = np.linspace(0,len(seq)-1,pts).astype(int)
        X.append(seq[idx].flatten())
    return np.vstack(X)
train_trajs = [np.array(a.traj) for a in agents]
X_train = preprocess(train_trajs)
model = OneClassSVM(kernel='rbf', nu=0.05, gamma='scale').fit(X_train)

# --- 2) MissingAgent 클래스 & 혼합 시뮬레이션 추가 ---
offroad_prob = 0.7
poi_change_prob = 0.7

class MissingAgent(Agent):
    def step(self, others):
        super().step(others)
        # random POI change
        if random.random()<poi_change_prob:
            (gx,gy),gw,gh = random.choice(goal_regions)
            self.goal = np.array([random.uniform(gx,gx+gw), random.uniform(gy,gy+gh)])
        # off-road jump
        if random.random()<offroad_prob:
            self.pos += np.random.randn(2)*0.5
            self.traj[-1]=self.pos.copy()

# simulate mixed
normals, missings = [], []
results=[]
agents_n, agents_m = [], []
for _ in range(num_steps):
    if random.random()<spawn_rate: agents_n.append(Agent(random.choice(start_regions)))
    if random.random()<spawn_rate/2: agents_m.append(MissingAgent(random.choice(start_regions)))
    # update normals
    for a in agents_n[:]:
        a.step(agents_n+agents_m)
        if not a.active or len(a.traj)>=50:
            results.append((np.array(a.traj), 1))
            agents_n.remove(a)
    # update missings
    for m in agents_m[:]:
        m.step(agents_n+agents_m)
        if not m.active or len(m.traj)>=50:
            results.append((np.array(m.traj), -1))
            agents_m.remove(m)

# --- 3) Predict & Visualize ---
fig, axs = plt.subplots(1,2,figsize=(12,6))
axs[0].set_title("Actual Normal")
axs[1].set_title("Actual Missing")
for traj,label in results:
    Xp = preprocess([traj])
    pred = model.predict(Xp)[0]
    ax = axs[0] if label==1 else axs[1]
    color = 'green' if pred==1 else 'red'
    ax.plot(traj[:,0],traj[:,1],color=color,alpha=0.7)
for ax in axs:
    ax.set_xlim(0,area_size); ax.set_ylim(0,area_size); ax.set_aspect('equal')
plt.show()
```
